# Problem 3: 
Use z-score normalization to standardize the values of the income attribute. Do not change the original
income attribute in the table.

```python
#Create z_income column and apply stats.zscore
data_zscores = data.copy()
data_zscores['income_zscore'] = (data_zscores.income - data_zscores.income.mean())/data_zscores.income.std()
data_zscores.head()
```

# Problem 4: 
Discretize the age attribute into 3 categories (corresponding to "young", "mid-age", and "old"). [Do not change the original age attribute in the table.]

```python
#Create age_bin column and create age bins
data_age_bins = data.copy()
data_age_bins['age_bin'] = pd.qcut(data_age_bins.age, [0, .33, .66, 1], labels = ['young', 'middle-aged', 'old'])
data_age_bins.head()
```

# Problem 5: 
Use Min-Max Normalization to transform the values of all numeric attributes (income, age, children) in
the original table (before the transforamtions in parts 3 and 4 above) onto the range 0.0-1.0.

```python
#Create numpy array for age, income, and children
data_min_max = data.copy()
data_min_max['income_minmax'] = (data_min_max.income - data_min_max.income.min())/(data_min_max.income.max() - data_min_max.income.min()) 
data_min_max['age_minmax'] = (data_min_max.age - data_min_max.age.min())/(data_min_max.age.max() - data_min_max.age.min()) 
data_min_max['children_minmax'] = (data_min_max.children - data_min_max.children.min())/(data_min_max.children.max() - data_min_max.children.min()) 
data_min_max.head()
```
